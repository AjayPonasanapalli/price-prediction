### Lets add some interesting features

```python
house_prices_all[['YearBuilt', 'YearRemodAdd', 'YrSold', 'MoSold']].hist(bins=12)
```

```python
plt.scatter(x=house_prices_train['YearBuilt'], y=house_prices_train['SalePrice'])
```

```python
# IR2 and IR3 don't appear that often, so just make a distinction between regular and irregular.
house_prices_all["IsRegularLotShape"] = (house_prices_all["LotShape"] == "Reg") * 1

# Most properties are level; bin the other possibilities together
# as "not level".
house_prices_all["IsLandLevel"] = (house_prices_all["LandContour"] == "Lvl") * 1

# Most land slopes are gentle; treat the others as "not gentle".
house_prices_all["IsLandSlopeGentle"] = (house_prices_all["LandSlope"] == "Gtl") * 1

# Most properties use standard circuit breakers.
house_prices_all["IsElectricalSBrkr"] = (house_prices_all["Electrical"] == "SBrkr") * 1

# About 2/3rd have an attached garage.
house_prices_all["IsGarageDetached"] = (house_prices_all["GarageType"] == "Detchd") * 1

# Most have a paved drive. Treat dirt/gravel and partial pavement
# as "not paved".
house_prices_all["IsPavedDrive"] = (house_prices_all["PavedDrive"] == "Y") * 1  

# If YearRemodAdd != YearBuilt, then a remodeling took place at some point.
house_prices_all["Remodeled"] = (house_prices_all["YearRemodAdd"] != house_prices_all["YearBuilt"]) * 1

# Did a remodeling happen in the year the house was sold?
house_prices_all["Sold_after_remodel"] = (house_prices_all["YearRemodAdd"] == house_prices_all["YrSold"]) * 1

# Was this house sold in the year it was built?
house_prices_all["Sold_right_away"] = (house_prices_all["YearBuilt"] == house_prices_all["YrSold"]) * 1

house_prices_all["Has2ndFloor"] = (house_prices_all["2ndFlrSF"] != 0) * 1
house_prices_all["HasMasVnr"] = (house_prices_all["MasVnrArea"] != 0) * 1
house_prices_all["HasWoodDeck"] = (house_prices_all["WoodDeckSF"] != 0) * 1
house_prices_all["HasOpenPorch"] = (house_prices_all["OpenPorchSF"] != 0) * 1
house_prices_all["HasEnclosedPorch"] = (house_prices_all["EnclosedPorch"] != 0) * 1
house_prices_all["Has3SsnPorch"] = (house_prices_all["3SsnPorch"] != 0) * 1
house_prices_all["HasScreenPorch"] = (house_prices_all["ScreenPorch"] != 0) * 1
house_prices_all["HasBasement"] = (house_prices_all["TotalBsmtSF"] != 0) * 1

# Months with the largest number of deals may be significant.
house_prices_all["HighSeason"] = house_prices_all["MoSold"].replace( 
    {1: 0, 2: 0, 3: 0, 4: 1, 5: 1, 6: 1, 7: 1, 8: 0, 9: 0, 10: 0, 11: 0, 12: 0})

house_prices_all["NewerDwelling"] = house_prices_all["MSSubClass"].replace(
    {20: 1, 30: 0, 40: 0, 45: 0,50: 0, 60: 1, 70: 0, 75: 0, 80: 0, 85: 0,
     90: 0, 120: 1, 150: 0, 160: 1, 180: 0, 190: 0}) 

house_prices_all["SaleCondition_PriceDown"] = house_prices_all.SaleCondition.replace(
        {'Abnorml': 1, 'Alloca': 1, 'AdjLand': 1, 'Family': 1, 'Normal': 0, 'Partial': 1})

# House completed before sale or not
house_prices_all["BoughtOffPlan"] = house_prices_all.SaleCondition.replace(
    {"Abnorml" : 0, "Alloca" : 0, "AdjLand" : 0, "Family" : 0, "Normal" : 0, "Partial" : 1})


area_cols = ['LotFrontage', 'LotArea', 'MasVnrArea',
             'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'GarageArea', 'WoodDeckSF', 
             'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'LowQualFinSF', 'PoolArea' ]
house_prices_all["TotalArea"] = house_prices_all[area_cols].sum(axis=1)

house_prices_all["TotalArea1st2nd"] = house_prices_all["1stFlrSF"] + house_prices_all["2ndFlrSF"]

house_prices_all["Age"] = 2010 - house_prices_all["YearBuilt"]
house_prices_all["TimeSinceSold"] = 2010 - house_prices_all["YrSold"]

house_prices_all["SeasonSold"] = house_prices_all["MoSold"].map({12:0, 1:0, 2:0, 3:1, 4:1, 5:1, 
                                              6:2, 7:2, 8:2, 9:3, 10:3, 11:3}).astype(int)

house_prices_all["YearsSinceRemodel"] = house_prices_all["YrSold"] - house_prices_all["YearRemodAdd"]

# Bin by neighborhood (a little arbitrarily). Values were computed by: 
# train_df["SalePrice"].groupby(train_df["Neighborhood"]).median().sort_values()
neighborhood_map = {
    "MeadowV" : 0,  #  88000
    "IDOTRR" : 1,   # 103000
    "BrDale" : 1,   # 106000
    "OldTown" : 1,  # 119000
    "Edwards" : 1,  # 119500
    "BrkSide" : 2,  # 124300
    "Sawyer" : 3,   # 135000
    "Blueste" : 3,  # 137500
    "SWISU" : 3,    # 139500
    "NAmes" : 3,    # 140000
    "NPkVill" : 4,  # 146000
    "Mitchel" : 5,  # 153500
    "SawyerW" : 6,  # 179900
    "Gilbert" : 7,  # 181000
    "NWAmes" : 7,   # 182900
    "Blmngtn" : 8,  # 191000
    "CollgCr" : 8,  # 197200
    "ClearCr" : 9,  # 200250
    "Crawfor" : 9,  # 200624
    "Veenker" : 10,  # 218000
    "Somerst" : 11,  # 225500
    "Timber" : 11,   # 228475
    "StoneBr" : 12,  # 278000
    "NoRidge" : 12,  # 290000
    "NridgHt" : 12,  # 315000
}

house_prices_all["NeighborhoodBin"] = house_prices_all["Neighborhood"].map(neighborhood_map)
```

```python
columns_to_poly_transform = [u'LotFrontage', u'LotArea', u'GrLivArea', 'TotalArea', 'TotalArea1st2nd']
```

```python
from sklearn import preprocessing
poly = preprocessing.PolynomialFeatures(2, include_bias=False) #include_bias=False means no intercept terms
temp_2nd_order = poly.fit_transform(house_prices_all[columns_to_poly_transform])
```

```python
temp_2nd_order.shape
```

```python
newFeatures = []
for entry in poly.powers_:
    newFeature = []
    for feat, coef in zip(columns_to_poly_transform, entry):
        if coef > 1:
            newFeature.append(feat+'**'+str(coef))
        elif coef == 1:
            newFeature.append(feat)
    if not newFeature:
        print(1) # If all powers are 0
    else:
        newFeatures.append(' * '.join(newFeature))
```

```python
newFeatures[0:20]
```

```python
house_2nd_order = pd.DataFrame(data=temp_2nd_order, columns=newFeatures)
house_2nd_order.head(1)
```

```python
remaining_columns = [x for x in house_prices_all.columns if x not in columns_to_poly_transform]
```

```python
house_2nd_order.shape, house_prices_all[remaining_columns].shape
```

```python
house_prices_all_2nd_poly = house_2nd_order.join(house_prices_all[remaining_columns], how='left')
house_prices_all_2nd_poly.head(2)
```
